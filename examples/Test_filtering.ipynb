```python
def cyclic3(i, shift):
    i0 = i - 1
    icyc0 = (i0 + shift) % 3
    return icyc0 + 1

def cyclic12(i, j):
    if (i == j):
        return None
    shifts = [cyclic3(i, 1), cyclic3(i, 2)]
    for s in shifts:
        if (j != s):
            return s
        
```
